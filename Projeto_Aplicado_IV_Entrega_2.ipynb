# **Referencial Teórico**

1. Principais Conceitos:

1.1 Queimadas:

As queimadas são incêndios não controlados que ocorrem em vegetações naturais ou agrícolas, e no Pantanal, essas queimadas são frequentemente causadas por práticas de manejo inadequadas e por condições climáticas extremas, como longos períodos de seca (SILVA et al., 2021). Além de destruir biodiversidade, elas têm implicações ambientais severas, incluindo a emissão de gases de efeito estufa, como o dióxido de carbono (CO2) e o carbono negro (CHEN et al., 2016).

1.2 Carbono negro:

O carbono negro é um tipo de partícula emitida principalmente pela combustão incompleta de biomassa e fósseis. Ele tem a capacidade de absorver calor na atmosfera, contribuindo para o aquecimento global (BOND et al., 2013). No contexto do Pantanal, o aumento dessas partículas pode influenciar a variação de temperatura regional, intensificando efeitos climáticos adversos (CURADO et al., 2023).

1.3 Séries temporais:

A análise de séries temporais consiste em examinar dados coletados ao longo do tempo para identificar padrões, tendências e sazonalidades. No contexto ambiental, essa técnica permite a previsão de fenômenos como a variação de temperatura e a incidência de queimadas, oferecendo uma ferramenta valiosa para políticas de mitigação e planejamento (SHUMWAY & STOFFER, 2017).


2. Trabalhos Correlacionados

Estudos anteriores sobre o impacto das queimadas em ecossistemas florestais e úmidos mostram relações diretas entre a ocorrência de incêndios e alterações climáticas regionais. Abaixo, detalhamos alguns desses estudos, com foco na metodologia aplicada.

2.1 Cochrane & Barber (2009) - Climate change, human land use and future fires in the Amazon:

Este estudo analisa a interação entre mudanças climáticas, uso da terra e queimadas na Amazônia, destacando como o desmatamento e a expansão agropecuária afetam a frequência e intensidade dos incêndios florestais. Utilizando dados de satélite sobre focos de calor e desmatamento, os autores mostram que a combinação entre a degradação do solo e as mudanças climáticas aumenta significativamente a probabilidade de queimadas. O estudo conclui que o desmatamento e as queimadas exacerbam o aquecimento regional, alertando para a necessidade urgente de políticas de manejo sustentável para mitigar esses impactos.

2.2 Dimri et al. (2020) - Time series analysis of climate variables using seasonal ARIMA approach:

Este estudo focou na modelagem de variáveis climáticas utilizando o modelo ARIMA (Autoregressive Integrated Moving Average) para prever parâmetros como precipitação e temperatura na Bacia do Rio Karkheh, no Irã. O objetivo foi compreender os padrões sazonais e as tendências de longo prazo dessas variáveis. A metodologia incluiu a análise de séries temporais de dados climáticos históricos, e os resultados mostraram que o ARIMA pode ser uma ferramenta eficaz para prever variáveis ambientais e apoiar a tomada de decisões relacionadas à gestão de recursos hídricos e mitigação de riscos climáticos.

2.3 Rasp et al. (2018) - Deep learning to represent subgrid processes in climate models:

Este estudo propôs uma abordagem inovadora utilizando redes neurais profundas (deep learning) para representar processos de subgrade em modelos climáticos. Subgrade refere-se a fenômenos de pequena escala, como convecção e turbulência, que afetam a modelagem climática em larga escala. A pesquisa demonstrou que o uso de redes neurais pode melhorar a precisão das simulações climáticas, permitindo a representação de processos complexos que geralmente são difíceis de modelar. O trabalho abriu novas possibilidades para a aplicação de aprendizado de máquina em climatologia, mostrando que essas técnicas podem superar os métodos tradicionais de parametrização usados em modelos climáticos.

A comparação será realizada com base em métricas como RMSE e MAE, buscando identificar qual abordagem oferece previsões mais precisas para o caso do pantanal.
