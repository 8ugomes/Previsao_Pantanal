# **Introdução**

O Pantanal, uma das maiores áreas úmidas do mundo, tem enfrentado um aumento expressivo na incidência de queimadas nas últimas décadas (WWF-BRASIL, 2024). Esses eventos não apenas destroem a rica biodiversidade local, mas também podem estar ligados a alterações climáticas regionais, como o aumento da concentração de carbono negro na atmosfera, que resulta em variações na temperatura média (CURADO et al., 2023). Esse cenário levanta preocupações sobre o equilíbrio climático e ambiental da região.

Diante dessa situação, este projeto busca entender a relação entre as queimadas e as mudanças de temperatura no Pantanal, explorando os impactos dessas interações no equilíbrio regional. A relevância deste estudo reside na necessidade urgente de compreender e mitigar os efeitos adversos das queimadas em um ecossistema tão importante. Com o aumento das queimadas e suas consequências climáticas, torna-se fundamental entender essa dinâmica para proteger a biodiversidade e o equilíbrio climático local. Além disso, a pesquisa oferece a oportunidade de gerar subsídios para políticas ambientais mais robustas e estratégias de mitigação, que podem beneficiar não só o Pantanal, mas também outras regiões afetadas por desafios semelhantes.

# **Referencial Teórico**

1. Principais Conceitos
As queimadas são incêndios não controlados que ocorrem em vegetações naturais ou agrícolas. No Pantanal, essas queimadas são frequentemente causadas por práticas de manejo inadequadas e por condições climáticas extremas, como longos períodos de seca. Segundo Silva et al. (2021), além de destruir a biodiversidade, as queimadas têm implicações ambientais severas, incluindo a emissão de gases de efeito estufa, como o dióxido de carbono (CO2) e o carbono negro.

O carbono negro é um tipo de partícula emitida principalmente pela combustão incompleta de biomassa e fósseis. Bond et al. (2013) destacam que ele tem a capacidade de absorver calor na atmosfera, contribuindo para o aquecimento global. No contexto do Pantanal, o aumento dessas partículas pode influenciar a variação de temperatura regional, intensificando os efeitos climáticos adversos, conforme apontado por Curado et al. (2023).

A análise de séries temporais consiste em examinar dados coletados ao longo do tempo para identificar padrões, tendências e sazonalidades. No contexto ambiental, essa técnica permite a previsão de fenômenos como a variação de temperatura e a incidência de queimadas, oferecendo uma ferramenta valiosa para políticas de mitigação e planejamento, segundo Shumway & Stoffer (2017).

2. Trabalhos Correlacionados
Estudos anteriores sobre o impacto das queimadas em ecossistemas florestais e úmidos mostram relações diretas entre a ocorrência de incêndios e alterações climáticas regionais. Abaixo, detalhamos alguns desses estudos, com foco na metodologia aplicada.

Cochrane & Barber (2009) exploraram a interação entre mudanças climáticas, uso da terra e queimadas na Amazônia. Eles destacaram como o desmatamento e a expansão agropecuária afetam a frequência e a intensidade dos incêndios florestais. Utilizando dados de satélite sobre focos de calor e desmatamento, os autores mostraram que a combinação entre a degradação do solo e as mudanças climáticas aumenta significativamente a probabilidade de queimadas. O estudo conclui que o desmatamento e as queimadas exacerbam o aquecimento regional, alertando para a necessidade urgente de políticas de manejo sustentável para mitigar esses impactos.

Dimri et al. (2020) focaram na modelagem de variáveis climáticas utilizando o modelo ARIMA (Autoregressive Integrated Moving Average) para prever parâmetros como precipitação e temperatura na Bacia do Rio Karkheh, no Irã. O objetivo foi compreender os padrões sazonais e as tendências de longo prazo dessas variáveis. A metodologia incluiu a análise de séries temporais de dados climáticos históricos, e os resultados mostraram que o ARIMA pode ser uma ferramenta eficaz para prever variáveis ambientais e apoiar a tomada de decisões relacionadas à gestão de recursos hídricos e mitigação de riscos climáticos.

Rasp et al. (2018) propuseram uma abordagem inovadora utilizando redes neurais profundas (deep learning) para representar processos de subgrade em modelos climáticos. Subgrade refere-se a fenômenos de pequena escala, como convecção e turbulência, que afetam a modelagem climática em larga escala. A pesquisa demonstrou que o uso de redes neurais pode melhorar a precisão das simulações climáticas, permitindo a representação de processos complexos que geralmente são difíceis de modelar. Este trabalho abriu novas possibilidades para a aplicação de aprendizado de máquina em climatologia, mostrando que essas técnicas podem superar os métodos tradicionais de parametrização usados em modelos climáticos.



# Pipeline da Solução
O projeto começa com a coleta de dados históricos de temperatura média e incidência de queimadas no Pantanal. Esses dados serão obtidos de fontes confiáveis, como agências meteorológicas e bases de dados ambientais. Em seguida, esses dados serão integrados para formar um conjunto abrangente que inclui variáveis relevantes, como temperatura, umidade, precipitação e focos de queimadas.

No pré-processamento, os dados serão limpos para tratar valores ausentes, duplicatas e outliers. Isso pode envolver técnicas como a interpolação para preencher valores ausentes e a remoção de dados irrelevantes. Em seguida, serão aplicadas técnicas de normalização ou padronização para garantir que os dados estejam na mesma escala, facilitando a modelagem. Além disso, serão criadas variáveis derivadas, como médias móveis ou índices de severidade de queimadas, para fornecer insights adicionais para os modelos.

A análise exploratória será realizada para entender as tendências e padrões nos dados. Serão criados gráficos e visualizações, como gráficos de linhas para séries temporais, histogramas para distribuição de dados e mapas de calor para correlações. Além disso, análises estatísticas serão conduzidas para identificar correlações e padrões sazonais nas séries temporais.

Para a modelagem, o conjunto de dados será dividido em conjuntos de treino e teste para validar a robustez dos modelos. O modelo ARIMA será implementado para capturar padrões lineares e sazonais nos dados, seguindo a abordagem de Dimri et al. (2020). Simultaneamente, o modelo LSTM (Long Short-Term Memory) será implementado para capturar padrões não lineares e dependências de longo prazo, conforme a abordagem de Rasp et al. (2018).

Na fase de avaliação, ambos os modelos serão treinados utilizando o conjunto de treino e suas previsões serão validadas no conjunto de teste. Essas métricas serão comparadas para determinar qual modelo oferece previsões mais precisas.

Após a avaliação, o modelo mais preciso será utilizado para realizar previsões de temperatura média com base na incidência de queimadas no Pantanal. As previsões serão visualizadas de forma clara e comparadas com os dados reais para destacar a precisão do modelo. Isso incluirá a criação de gráficos de linhas, diagramas de dispersão e mapas de calor.

Por fim, todo o processo será documentado em um relatório final, incluindo a metodologia, resultados das avaliações e conclusões. O relatório também incluirá recomendações para políticas de mitigação e planejamento com base nas previsões obtidas. Este pipeline proporciona uma abordagem estruturada e sistemática para prever variações de temperatura no Pantanal com base na incidência de queimadas, garantindo que cada etapa do processo seja cuidadosamente executada e documentada.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Carregar datasets
data_focos = pd.read_csv('/content/historico_bioma_pantanal.csv',sep=';')
data_temp = pd.read_csv('/content/bquxjob_3ea7ff59_192c40cd453.csv')

# Converter colunas de data e hora para datetime
data_temp['datetime'] = pd.to_datetime(data_temp['data'] + ' ' + data_temp['hora'], errors='coerce')

# Tratamento de valores ausentes com interpolação
data_temp['temperatura_max'] = data_temp['temperatura_max'].interpolate()
data_temp['temperatura_min'] = data_temp['temperatura_min'].interpolate()

# Verificar a distribuição das temperaturas máximas e mínimas
plt.figure(figsize=(12, 5))
sns.histplot(data_temp['temperatura_max'], kde=True, color='red', label='Temperatura Máxima')
sns.histplot(data_temp['temperatura_min'], kde=True, color='blue', label='Temperatura Mínima')
plt.title('Distribuição das Temperaturas Máximas e Mínimas')
plt.legend()
plt.show()

# Gráficos de linha para tendências ao longo do tempo
plt.figure(figsize=(14, 6))
data_temp.set_index('datetime')[['temperatura_max', 'temperatura_min']].plot(alpha=0.7)
plt.title('Tendências das Temperaturas Máximas e Mínimas ao Longo do Tempo')
plt.ylabel('Temperatura (°C)')
plt.show()

# Resumo estatístico após o preenchimento de valores ausentes
print(data_temp[['temperatura_max', 'temperatura_min']].describe())
```

# **Modelo base**

Como será feita a comparação do modelo ARIMA e LSTM como já demonstrado através da literatura, através das métricas de RMSE e MAE, nessa sessão será apresentado o desenvolvimento dos modelos bem como suas métricas de avaliação.

```python
#Modelo ARIMA
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

# Dicionário para mapear os meses de português para números
mes_map = {
    'Janeiro': '01', 'Fevereiro': '02', 'Março': '03', 'Abril': '04',
    'Maio': '05', 'Junho': '06', 'Julho': '07', 'Agosto': '08',
    'Setembro': '09', 'Outubro': '10', 'Novembro': '11', 'Dezembro': '12'
}

# Converter o dataset de focos para uma série temporal mensal
data_focos_melted = data_focos.melt(id_vars=['Ano'],
                                    value_vars=['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho',
                                                'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro'],
                                    var_name='Mes', value_name='Focos')

# Substituir o nome do mês pelo número correspondente
data_focos_melted['Mes'] = data_focos_melted['Mes'].map(mes_map)

# Criar a coluna 'date' com ano e número do mês
data_focos_melted['date'] = pd.to_datetime(data_focos_melted['Ano'].astype(str) + '-' +
                                           data_focos_melted['Mes'] + '-01', format='%Y-%m-%d')

# Ordenar por data e redefinir o índice
data_focos_melted = data_focos_melted.sort_values('date').set_index('date')

# Preparar o dataset de temperatura
data_temp['date'] = pd.to_datetime(data_temp['ano'].astype(str) + '-' + data_temp['mes'].astype(str) + '-01')
data_temp = data_temp.set_index('date')

# Consolidar os datasets de focos e temperatura
data_combined = pd.merge(data_focos_melted[['Focos']], data_temp[['temperatura_max']],
                         left_index=True, right_index=True, how='inner')
```

A Análise de Autocorrelação (ACF) e Autocorrelação Parcial (PACF) são ferramentas importantes para ajudar a determinar os parâmetros p, d e q do modelo ARIMA.

```python
# Análise de autocorrelação para ARIMA
fig, axes = plt.subplots(1, 2, figsize=(16, 4))
plot_acf(data_combined['temperatura_max'].dropna(), ax=axes[0])
plot_pacf(data_combined['temperatura_max'].dropna(), ax=axes[1])
plt.show()
```

```python
# Parâmetros iniciais para o modelo ARIMA
p, d, q = 1, 1, 1

# Aplicação do modelo ARIMA
model = ARIMA(data_combined['temperatura_max'], exog=data_combined['Focos'], order=(p, d, q))
arima_resultado = model.fit()

# Exibir resumo do modelo
print(arima_resultado.summary())
```

```python
# Previsão e visualização
data_combined['previsao'] = arima_resultado.predict(start=0, end=len(data_combined)-1, exog=data_combined['Focos'])
plt.figure(figsize=(12, 6))
plt.plot(data_combined['temperatura_max'], label='Temperatura Observada', color='blue')
plt.plot(data_combined['previsao'], label='Previsão ARIMA', color='red', linestyle='--')
plt.title('Previsão de Temperatura com Modelo ARIMA e Focos de Incêndio')
plt.xlabel('Ano')
plt.ylabel('Temperatura Máxima (°C)')
plt.legend()
plt.show()
```

```python
# Verificar a presença de campos nulos no dataset combinado
print("Quantidade de NaNs em 'temperatura_max':", data_combined['temperatura_max'].isna().sum())
print("Quantidade de NaNs em 'previsao':", data_combined['previsao'].isna().sum())
```

```python
# Remover linhas com NaNs nas colunas de interesse para o cálculo de métricas
data_combined_clean = data_combined.dropna(subset=['temperatura_max', 'previsao'])

# Calcular RMSE e MAE entre os valores observados e previstos (após remover NaNs)
from sklearn.metrics import mean_squared_error, mean_absolute_error

rmse = np.sqrt(mean_squared_error(data_combined_clean['temperatura_max'], data_combined_clean['previsao']))
mae = mean_absolute_error(data_combined_clean['temperatura_max'], data_combined_clean['previsao'])

print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"Mean Absolute Error (MAE): {mae}")
```

```python
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout


# Verificar dados iniciais
print(data_combined.isnull().sum())

# Remover linhas com NaN antes de qualquer transformação
data_combined = data_combined.dropna()

# Conversão de tipos
data_combined['Focos'] = data_combined['Focos'].astype(float)

# Normalizar os dados
scaler = MinMaxScaler(feature_range=(0, 1))
data_scaled = scaler.fit_transform(data_combined[['temperatura_max', 'Focos']])

# Criar dados para LSTM
def create_dataset(data, time_step=1):
    X, y = [], []
    for i in range(len(data) - time_step - 1):
        a = data[i:(i + time_step), :]
        X.append(a)
        y.append(data[i + time_step, 0])
    return np.array(X), np.array(y)

# Definir o número de passos de tempo
time_step = 12
X, y = create_dataset(data_scaled, time_step)

# Dividir os dados em conjuntos de treinamento e teste
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Reshape para [samples, time steps, features] para LSTM
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 2)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 2)

# Construção do modelo LSTM
model_lstm = Sequential()
model_lstm.add(LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], 2)))
model_lstm.add(Dropout(0.2))
model_lstm.add(LSTM(50, return_sequences=False))
model_lstm.add(Dropout(0.2))
model_lstm.add(Dense(1))

# Compilar o modelo
model_lstm.compile(optimizer='adam', loss='mean_squared_error')

# Treinar o modelo
model_lstm.fit(X_train, y_train, epochs=10, batch_size=32, verbose=1)

# Previsão com o modelo LSTM
lstm_pred = model_lstm.predict(X_test)

# Verificar se lstm_pred contém NaN
print(np.isnan(lstm_pred).sum())

# Inversão da normalização
lstm_pred = scaler.inverse_transform(np.concatenate((lstm_pred, np.zeros((lstm_pred.shape[0], 1))), axis=1))[:, 0]

# Visualização da previsão LSTM
plt.figure(figsize=(12, 6))
plt.plot(data_combined['temperatura_max'], label='Temperatura Observada', color='blue')
plt.plot(data_combined.index[-len(lstm_pred):], lstm_pred, label='Previsão LSTM', color='green', linestyle='--')
plt.title('Previsão de Temperatura com Modelo LSTM')
plt.xlabel('Ano')
plt.ylabel('Temperatura Máxima (°C)')
plt.legend()
plt.show()

# Calcular RMSE e MAE para LSTM
rmse_lstm = np.sqrt(mean_squared_error(y_test, lstm_pred))
mae_lstm = mean_absolute_error(y_test, lstm_pred)

print(f"Root Mean Squared Error (RMSE) LSTM: {rmse_lstm}")
print(f"Mean Absolute Error (MAE) LSTM: {mae_lstm}")
```

Em uma primeira análise, o método ARIMA apresentou melhores resultados em relação a MAE e RMSE. A fim de evoluir os modelos, sugere-se que os valores dos parâmetros p,d e d de ARIMA sejam alterados em função do aprofundamento da análise de autocorrelação. Já para o modelo LSTM sugere-se o aumento do número de épocas.
