# **Introdução**

O Pantanal, uma das maiores áreas úmidas do mundo, enfrenta um aumento significativo na incidência de queimadas nas últimas décadas (WWF-BRASIL, 2024). Esses eventos não apenas destroem a rica biodiversidade local, mas também podem estar associados a alterações climáticas regionais através do aumento da concentração de carbono negro no ar, resultando em variações na temperatura média (CURADO et al., 2023). Diante desse cenário, o projeto busca entender a relação entre as queimadas e a mudança de temperatura no Pantanal, explorando seus impactos no equilíbrio climático e ambiental da região. Através da análise de séries temporais, o objetivo é prever o impacto na temperatura média com base na incidência de focos de queimadas, fornecendo insights para a formulação de estratégias de mitigação e políticas ambientais mais eficazes.



# **Motivações e justificativa**

A relevância do tema reside na urgente necessidade de compreender e mitigar os efeitos adversos das queimadas no Pantanal, uma das maiores áreas úmidas do mundo. As queimadas não só destroem a rica biodiversidade local, mas também podem desencadear alterações climáticas regionais, como a variação na temperatura média. Com o aumento da incidência de queimadas nas últimas décadas, entender essa relação torna-se essencial para proteger o equilíbrio climático e ambiental da região. Além disso, a pesquisa pode fornecer insights valiosos para a formulação de políticas ambientais mais eficazes e estratégias de mitigação. Os benefícios envolvem a preservação do ecossistema do Pantanal, a proteção da biodiversidade e a melhoria das condições climáticas locais, além de servir como um modelo para outras regiões enfrentando problemas similares.

# **Objetivo**

O objetivo deste projeto é investigar a relação entre a incidência de queimadas e a variação na temperatura média no Pantanal. Utilizando a análise de séries temporais, pretende-se prever o impacto das queimadas na temperatura média da região. A partir desses dados, busca-se fornecer subsídios para a formulação de estratégias de mitigação mais eficazes e políticas ambientais que possam preservar o equilíbrio climático e ambiental do Pantanal. O projeto almeja, assim, contribuir para a conservação de um dos ecossistemas mais importantes do mundo e fornecer um modelo de análise aplicável a outras regiões afetadas por problemas similares.
